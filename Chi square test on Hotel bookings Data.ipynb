```python
df1 = df[['hotel','distribution_channel','deposit_type','customer_type','is_repeated_guest','reservation_status','is_canceled']]
```

```python
df1.isna().sum()
```

```python
df1
```

```python
import numpy as np
```

```python
df1['hotel'] = np.where(df1['hotel']=='Resort Hotel',1,0)
```

```python
df1
```

```python
#lets perform encoding on other columns
```

```python
ordinal_label = {k:i for i,k in enumerate(df1['distribution_channel'].unique(),0)}
```

```python
ordinal_label
```

```python
df1['distribution_channel'] = df1['distribution_channel'].map(ordinal_label)
```

```python
df1
```

```python
ordinal_label = {k:i for i,k in enumerate(df1['deposit_type'].unique(),0)}
df1['deposit_type'] = df1['deposit_type'].map(ordinal_label)
```

```python
ordinal_label = {k:i for i,k in enumerate(df1['customer_type'].unique(),0)}
df1['customer_type'] = df1['customer_type'].map(ordinal_label)
```

```python
ordinal_label = {k:i for i,k in enumerate(df1['reservation_status'].unique(),0)}
df1['reservation_status'] = df1['reservation_status'].map(ordinal_label)
```

```python
df1
```

```python
# Train test split is usually done to avoid overfitting
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test =train_test_split(df1[['hotel','distribution_channel','deposit_type','customer_type','is_repeated_guest','reservation_status']],
                                               df1['is_canceled'],test_size = 0.3,random_state =100)
```

```python
X_train.isna().sum()
```

```python
# Perform chi2 test -- it returns two values f-score & p value
from sklearn.feature_selection import chi2
f_p_values =chi2(X_train,y_train)
```

```python
f_p_values
```

```python
p_values = pd.Series(f_p_values[1])
p_values.index = X_train.columns
p_values
```

```python
p_values.sort_index(ascending=False)
```

```python
f_values = pd.Series(f_p_values[0])
f_values.index = X_train.columns
f_values
```
